```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.signal import lfilter
from dspmini import *
```

```python
# Seal importada de csv
data = np.loadtxt("../data/senial_14253.csv", delimiter=",")
nx = data[:, 0].astype(int)
x = data[:, 1]
pairs = list(zip(nx, x))

# INPUT: Defino en tiempo x[n] y frecuencia X(e^j2s)
nx, x = signal_from_pairs(pairs)
sX, X = tftd_numeric(nx, x, N_window=None)
plot_grid_time_freq(nx, x, sX, X, title_prefix="")
```

```python
# S3 System
b = [0.25, 0.25]   # numerador   -> 0.25 x[n] + 0.25 x[n-1]
a = [1.0, 0.5]    # denominador -> 1.00 y[n] + 0.50 y[n-1]

# y3[n] Output Signal
y = lfilter(b, a, x)
ny = np.arange(len(y))
pairs_y = list(zip(ny, y))

# OUTPUT: Defino en tiempo y3[n] y frecuencia Y3(e^j2s)
ny, y = signal_from_pairs(pairs_y)
sY, Y = tftd_numeric(ny, y, N_window=None)

plot_grid_time_freq(ny, y, sY, Y, title_prefix="")
```

```python
# S3 System
b = [0.25, -0.25]   # numerador   -> 0.25 x[n] - 0.25 x[n-1]
a = [1.0, -0.5]    # denominador -> 1.00 y[n] - 0.50 y[n-1]

# y3[n] Output Signal
y = lfilter(b, a, x)
ny = np.arange(len(y))
pairs_y = list(zip(ny, y))

# OUTPUT: Defino en tiempo y3[n] y frecuencia Y3(e^j2s)
ny, y = signal_from_pairs(pairs_y)
sY, Y = tftd_numeric(ny, y, N_window=None)

plot_grid_time_freq(ny, y, sY, Y, title_prefix="")
```
