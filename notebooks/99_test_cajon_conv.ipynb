```python
import numpy as np
from dspmini import (
    rect_centered, signal_from_pairs, conv_discrete,
    tftd_numeric, stem_discrete, plot_mag, plot_phase
)

# Seal x[n]: cajn N=7 centrado
nx, x = rect_centered(7)
stem_discrete(nx, x, "x[n] (N=7)")

# Sistema h[n]: cajn N=7 centrado
nh, h = rect_centered(7)
stem_discrete(nh, h, "h[n] (N=7)")

# TFTD numrica
s, X = tftd_numeric(nx, x)           # sin ventana adicional (seal finita)
plot_mag(s, X, "X(e^{j2s})")

s, H = tftd_numeric(nh, h)
plot_mag(s, H, "H(e^{j2s})")

# Convolucin
ny, y = conv_discrete(nx, x, nh, h)
stem_discrete(ny, y, "y[n] = (x*h)[n]")
```

```python
def expo_real(a, n_min=0, n_max=80):
    n = np.arange(n_min, n_max+1)
    x = a**n
    return n, x

n2, x2 = expo_real(0.9)
s2, X2 = tftd_numeric(n2, x2, N_window=15)
plot_mag(s2, X2, "a^n u[n]")
```
