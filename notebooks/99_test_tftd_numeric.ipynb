```python
import numpy as np
from dspmini import (
    rect_centered, signal_from_pairs, conv_discrete,
    tftd_numeric, stem_discrete, plot_mag, plot_phase
)

pairs = [(-4, 0.5), (-3, 1.0), (0, 2.0), (2, -1.0), (3, 0.5)]
n1, x1 = signal_from_pairs(pairs)
stem_discrete(n1, x1, "x1[n] desde pares")
s1, X1 = tftd_numeric(n1, x1, N_window=51)  # ejemplo aplicando cajn para truncar
plot_mag(s1, X1, "X1(e^{j2s})")
```

```python
pairs = [(-3, 0), (-3, 0), (0, 1.0), (2, 0), (3, 0)]
n2, x2 = signal_from_pairs(pairs)
stem_discrete(n2, x2, "x2[n] desde pares")
s2, X2 = tftd_numeric(n2, x2, N_window=51)  # ejemplo aplicando cajn para truncar
plot_mag(s2, X2, "X2(e^{j2s})")
```

```python
f0 = 0.125     # frecuencia normalizada
phi = 0        # fase en radianes
n_min, n_max = -20, 20   # soporte discreto elegido

# Muestreo
n = np.arange(n_min, n_max+1, dtype=int)
x = np.cos(2*np.pi*f0*n + phi)

# Construccin como lista de pares (n, valor)
pairs = list(zip(n, x))
nx, x3 = signal_from_pairs(pairs)

# Graficar tiempo
stem_discrete(nx, x3, "x3[n] = cos(2 f0 n + )")

# TFTD numrica (se aplica cajn ancho para aproximar mejor la suma si se desea)
s3, X3 = tftd_numeric(nx, x3, N_window=101)

# Mdulo y fase de X(e^{j2s})
plot_mag(s3, X3, "X3(e^{j2s})")
```
